### Simule dois sinais senoidais com frequências 5 e 500 Hz e amplitudes 0.5 e 1, cuja frequencia de amostragem é de 5 kHz. Crie um vetor n 600 e responda quantos ciclos de senoide serão representados nos dois sinais senoidais.

```python
import numpy as np 
import matplotlib.pylab as plt
from scipy.fft import fft, fftfreq
```

```python
fcos1 = 500
fcos2 = 5000
fs = 50000

n = np.arange(0,int(6*(fs/fcos1)),1)

cos5 = np.cos((2*np.pi*fcos1/fs)*n)*.5
cos500 = np.cos((2*np.pi*fcos2/fs)*n)
```

```python
plt.figure(figsize=(12,12))
plt.subplot(2,1,1)
plt.plot(n,cos5,'b')
#plt.axis([0,n[-1],-1,1])

plt.subplot(2,1,2)
plt.plot(n,cos500,'b')
#plt.axis([0,10*5,-1,1])
```

# Realize a multiplicação entre os dois sinais. Calcule a transformada de fourier utilizando o comando fft e represente a sua resposta para os tres sinais

```python
cosAM = cos5*cos500
plt.figure(figsize=(12,12))
plt.plot(n,cosAM)
```

```python
N = 2**16
FFTCOS5 = fft(cos5*np.hanning(cos5.size),N)
FFTCOS500 = fft(cos500,N)
FFTCOSAM = fft(cosAM,N)
freq_axisx = fftfreq(N,fs)
```

```python
plt.figure(figsize=(12,12))
plt.subplot(1,3,1)
plt.plot(freq_axisx,np.abs(FFTCOS5))


plt.subplot(1,3,2)
plt.plot(freq_axisx,np.abs(FFTCOS500))
plt.subplot(1,3,3)
plt.plot(freq_axisx,np.abs(FFTCOSAM))
```
