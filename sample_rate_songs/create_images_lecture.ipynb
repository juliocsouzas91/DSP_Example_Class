```python
import matplotlib.pylab as plt
import numpy as np 
from plot_setup import *
define_plots()
```

```python
#Declaração das variáveis utilizadas
fc = 440 #frequência do sinal
fs = 8800000 #frequência de amostragem
periodo_amostral = 1/fs
tempo_amostrando = 1/440
n_samples = np.arange(0,fs*tempo_amostrando, 1)
```

```python
#Criando e ilustrndo o sinal senoidal
sinal_senoidal = np.sin(2*np.pi*fc/fs*n_samples)
t = (1/fs)*n_samples
conversor_escala_tempo = 1000 
plt.plot(t*conversor_escala_tempo,sinal_senoidal,'k')
plt.axis([0, (1/fc)*conversor_escala_tempo , -1.01, 1.01])
plt.xlabel('Tempo (ms)')
plt.ylabel('Amplitude')
plt.savefig("sinal_input.svg",bbox_inches='tight')
```
